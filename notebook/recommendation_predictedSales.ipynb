```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/zoneAnalysis1111.csv",index_col=0)
df
```

```python
df_best = df.sort_values(by=['predicted_sales'],ascending=False).head(3)
df_worst = df.sort_values(by=['predicted_sales'],ascending=True).head(3)
```

```python
df_best
```

```python
import folium

m = folium.Map(location=[22.3193, 114.1694], zoom_start=11)

# create a rectangle object and add to map
for row,data in df_best.iterrows():
    folium.Rectangle([(data['lat2'],data['lon2']), (data['lat1'],data['lon1'])],color='green').add_to(m)

# create a rectangle object and add to map
for row,data in df_worst.iterrows():
    folium.Rectangle([(data['lat2'],data['lon2']), (data['lat1'],data['lon1'])],color='red').add_to(m)

# folium.Rectangle([(28.6471948,76.9531796), (19.0821978,72.7411)]).add_to(m)

# save the map object as a html
m
```
