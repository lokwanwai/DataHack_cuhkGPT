```python
import pandas as pd
import json
import googlemaps
import time
import folium
```

```python
with open("../config/googleAPI.json") as f:
    config = json.load(f)
    GOOGLE_API_KEY = config["GOOGLE_API_KEY"]
```

```python
df_wellcome = pd.read_csv("../data/processed/wellcomeMap.csv")
df_wellcome
```

```python
# the first of most common catorgory in types
catorgory = list(eval(df_wellcome["types"].value_counts(ascending=False).index[0]))[0]
catorgory
```

```python
import googlemaps
from datetime import datetime

# set up API key and client
gmaps = googlemaps.Client(key=GOOGLE_API_KEY)
```

```python
def search_all(search_query):

    df = pd.DataFrame()

    # Lantau Island
    results = gmaps.places(
        type=search_query,
        location=(22.271728, 113.953225),
        radius=5000,
        )
    df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
    print("Total number of results: ", len(df))

    while "next_page_token" in results:
        time.sleep(2)
        results = gmaps.places(
            type=search_query,
            location=(22.271728, 113.953225),
            radius=5000,
            page_token=results["next_page_token"],
            )
        df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
        print("Total number of results: ", len(df))


    # Hong Kong Island
    results = gmaps.places(
        type=search_query,
        location=(22.267006, 114.185718),
        radius=5000,
        )
    df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
    print("Total number of results: ", len(df))

    while "next_page_token" in results:
        time.sleep(2)
        results = gmaps.places(
            type=search_query,
            location=(22.267006, 114.185718),
            radius=5000,
            page_token=results["next_page_token"],
            )
        df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
        print("Total number of results: ", len(df))
    
    # Tin Shui Wai
    results = gmaps.places(
        type=search_query,
        location=(22.420820, 113.988650),
        radius=5000,
        )
    df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
    print("Total number of results: ", len(df))

    while "next_page_token" in results:
        time.sleep(2)
        results = gmaps.places(
            type=search_query,
            location=(22.420820, 113.988650),
            radius=5000,
            page_token=results["next_page_token"],
            )
        df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
        print("Total number of results: ", len(df))

    # Tai Po
    results = gmaps.places(
        type=search_query,
        location=(22.434396, 114.190205),
        radius=5000,
        )
    df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
    print("Total number of results: ", len(df))

    while "next_page_token" in results:
        time.sleep(2)
        results = gmaps.places(
            type=search_query,
            location=(22.434396, 114.190205),
            radius=5000,
            page_token=results["next_page_token"],
            )
        df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
        print("Total number of results: ", len(df))

    # Kowloon
    results = gmaps.places(
        type=search_query,
        location=(22.331288, 114.204035),
        radius=5000,
        )
    df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
    print("Total number of results: ", len(df))

    while "next_page_token" in results:
        time.sleep(2)
        results = gmaps.places(
            type=search_query,
            location=(22.331288, 114.204035),
            radius=5000,
            page_token=results["next_page_token"],
            )
        df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
        print("Total number of results: ", len(df))
    
    # Tsuen Wan
    results = gmaps.places(
        type=search_query,
        location=(22.364287, 114.122800),
        radius=5000,
        )
    df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
    print("Total number of results: ", len(df))

    while "next_page_token" in results:
        time.sleep(2)
        results = gmaps.places(
            type=search_query,
            location=(22.364287, 114.122800),
            radius=5000,
            page_token=results["next_page_token"],
            )
        df = pd.concat([df, pd.DataFrame(results["results"])], axis=0)
        print("Total number of results: ", len(df))

    return df
```

```python
df = search_all(catorgory)
```

```python
df.drop_duplicates(subset="place_id", inplace=True)
df_filtered = df[[ 'name', 'formatted_address','place_id', 'rating', 'user_ratings_total','types']]
df_filtered.loc[:,'latitude'] = df['geometry'].apply(lambda x: x['location']['lat'])
df_filtered.loc[:,'longitude'] = df['geometry'].apply(lambda x: x['location']['lng'])
df_filtered.reset_index(drop=True, inplace=True)
df_filtered.dropna(subset=['name','latitude','longitude'], inplace=True)
df_filtered
```

```python
df_filtered.to_csv("../data/processed/competitor.csv", index=False)
```

```python

# create a map centered at Hong Kong
m = folium.Map(location=[22.3193, 114.1694], zoom_start=11)

# iterate through the rows of the dataframe and add markers to the map
for index, row in df_filtered.iterrows():
    name = row['name']
    lat = row['latitude']
    lon = row['longitude']
    popup = folium.Popup(name)
    marker = folium.Marker(location=[lat, lon], popup=popup)
    marker.add_to(m)

# display the map
m
```

```python
m.save("../data/processed/competitor.html")
```
