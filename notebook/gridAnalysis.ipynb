```python
import pandas as pd

# assume df_zone is a dataframe with columns lat1, lat2, lon1, lon2 that define your zones
# assume df_wellcome is a dataframe with columns latitude and longitude that define the locations of the shops

# create an empty list to store the number of shops in each zone
num_shops = []
rating_list = []
num_rating_list = []

# loop through each zone in df_zone
for i in range(len(df_zone)):
    # extract the boundaries of the current zone
    lat1, lat2, lon1, lon2 = df_zone.iloc[i][["lat1", "lat2", "lon1", "lon2"]]

    # count the number of shops in the current zone by checking if their latitude and longitude are within the zone boundaries
    num_shops_in_zone = (
        (df_wellcome["latitude"] >= lat1)
        & (df_wellcome["latitude"] <= lat2)
        & (df_wellcome["longitude"] >= lon1)
        & (df_wellcome["longitude"] <= lon2)
    ).sum()

    rating = df_wellcome[
        (df_wellcome["latitude"] >= lat1)
        & (df_wellcome["latitude"] <= lat2)
        & (df_wellcome["longitude"] >= lon1)
        & (df_wellcome["longitude"] <= lon2)
    ]["rating"].mean()

    rating_total = df_wellcome[
        (df_wellcome["latitude"] >= lat1)
        & (df_wellcome["latitude"] <= lat2)
        & (df_wellcome["longitude"] >= lon1)
        & (df_wellcome["longitude"] <= lon2)
    ]["user_ratings_total"].mean()

    # append the number of shops in the current zone to the list
    num_shops.append(num_shops_in_zone)
    rating_list.append(rating)
    num_rating_list.append(rating_total)



# add the num_shops list as a new column to df_zone
df_zone["Number_of_shop"] = num_shops
df_zone["ratingMean"] = rating_list
df_zone['Number_of_rating_Mean'] = num_rating_list
```

# busStop

```python
def get_num_of_target_place(df_zone,df_target):
    num = []
    for i in range(len(df_zone)):
        lat1, lat2, lon1, lon2 = df_zone.iloc[i][['lat1', 'lat2', 'lon1', 'lon2']]
        num_busStop_in_zone = ((df_target['latitude'] >= lat1) & (df_target['latitude'] <= lat2) & (df_target['longitude'] >= lon1) & (df_target['longitude'] <= lon2)).sum()
        num.append(num_busStop_in_zone)
    return num
```

```python
df_zone['Number_of_busStop'] = get_num_of_target_place(df_zone,df_busStop)
```

# Mtr Station

```python
df_zone['Number_of_Mtr'] = get_num_of_target_place(df_zone,df_mtr)
df_zone
```

```python
df_competitor = pd.read_csv("../data/processed/competitor.csv")
df_competitor
```

```python
df_zone['Number_of_competitor'] = get_num_of_target_place(df_zone,df_competitor)
```
