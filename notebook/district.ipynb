```python
import pandas as pd
import numpy as np
```

```python
df_district = pd.read_csv('../data/raw/district.csv')
df_district
```

```python
df_pop = pd.read_csv('../data/raw/district_pop.csv')
df_pop = df_pop[['district', 'Total', 'Average household sie', 'Median income']]
df_pop[['Total', 'Average household sie', 'Median income']] =df_pop[['Total', 'Average household sie', 'Median income']].replace({',': ''}, regex=True)
df_pop[['Total', 'Average household sie', 'Median income']] = df_pop[['Total', 'Average household sie', 'Median income']].astype(float)
df_pop
```

```python
df = pd.merge(df_district, df_pop, on='district')
df
```

```python
df.to_csv("../data/processed/districtAll.csv", index=False)
```
